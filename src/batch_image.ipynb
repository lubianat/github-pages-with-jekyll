```python
import pytesseract as ocr

from PIL import Image


phrase = ocr.image_to_string(Image.open('../data/temp.jpeg'), lang='por')
print(phrase)
```

Encontrando todas as datas:

```python
import re
from icalendar import Calendar, Event
import datetime as dt

cal = Calendar()




pattern_hora = r'([0-9][0-9])h'
hora = re.findall(pattern_hora, phrase)
if hora:
    hora = int(hora[0])
    print(hora)


pattern_minuto = r'h([0-9][0-9])'
minuto = re.findall(pattern_minuto, phrase)
if minuto:
    minuto = int(minuto[0])
    print(minuto)


ano = 2019
    


pattern_local = r'(Sala .*)'
local = re.findall(pattern_local, phrase)
if local:
    local = local[0]
    print(local)



#Get the beggining of each date
beg = 0
for index, m in enumerate(re.finditer(pattern_dia, phrase)):
    print(index)
    if index == 0:
        continue
    end = m.span()[0]
    event = phrase[beg:end]
    print(event)
    
    #######
    pattern_dia = r'([0-9][0-9])/'
    dia = re.findall(pattern_dia, event)
    if dia:
        dia = int(dia[0])
        print(type(dia))
    print(dia)
    
    #######
    pattern_mes = r'/([0-9][0-9])'
    mes = re.findall(pattern_mes, event)
    if mes:
        mes = int(mes[0])
    mes = mes
    print(mes) 
    
    ####### 
    pattern_titulo = r'[—-] (.*?[\?\.])'
    titulo = re.search(pattern_titulo, event, flags=re.S)
    if (titulo):
        titulo = titulo.group()[2:]
        print(titulo) 

    
    
    #######
    
    event_start = dt.datetime(ano, mes, dia, hora, minuto, 0)
    event_end = dt.datetime(ano, mes, dia, hora + 1, minuto, 0)
    
    evento = Event()
    evento.add('dtstart', event_start)
    evento.add('dtend', event_end)
    evento.add('summary', titulo )
    evento.add('location', local)
    evento.add('description', titulo + "\n" + "part of \n" +phrase)
    cal.add_component(evento)
    
    #######
    beg = end

event = phrase[beg:-1]
#######
pattern_dia = r'([0-3][0-9])/'
dia = re.findall(pattern_dia, event)
if dia:
    dia = int(dia[0])
print(dia)
    
#######
pattern_mes = r'/([0-1][0-9])'
mes = re.findall(pattern_mes, event)
if mes:
    mes = int(mes[0])
mes = mes
print(mes) 

####### 
pattern_titulo = r'[—-] (.*?[\?\.])'
titulo = re.search(pattern_titulo, event, flags=re.S)
if (titulo):
    titulo = titulo.group()[2:]
    print(titulo) 

event_start = dt.datetime(ano, mes, dia, hora, minuto, 0)
event_end = dt.datetime(ano, mes, dia, hora + 1, minuto, 0)
    
evento = Event()
evento.add('dtstart', event_start)
evento.add('dtend', event_end)
evento.add('summary', titulo )
evento.add('location', local)
evento.add('description', titulo + "\n" + "part of \n" +phrase)
cal.add_component(evento)
```

```python
import tempfile, os
directory = tempfile.mkdtemp()
f = open(os.path.join(directory, 'example.ics'), 'wb')
f.write(cal.to_ical())
f.close()
```
